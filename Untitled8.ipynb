```python
# Importar Bibliotecas:

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar os Dados:

# DataFrame
dados = {
    'Dias': ['Domingo', 'Segunda', 'Tera', 'Quarta', 'Quinta', 'Sexta', 'Sbado'],
    'Horas Trabalhadas': [8, 7, 8, 6, 7, 8, 5],
    'Bugs Corrigidos': [3, 1, 2, 0, 4, 1, 0],
    'Tarefas Concludas': [5, 4, 6, 3, 7, 5, 2]
}

df = pd.DataFrame(dados)
# Anlise Exploratria:

# Visualizando as primeiras linhas do DataFrame
df.head()
# Estatsticas Descritivas:
# Calculando estatsticas descritivas
df.describe()
# Visualizaes:
# Grfico de linhas para Horas Trabalhadas, Bugs Corrigidos e Tarefas Concludas
plt.figure(figsize=(10, 6))
sns.lineplot(x='Dias', y='Horas Trabalhadas', data=df, label='Horas Trabalhadas')
sns.lineplot(x='Dias', y='Bugs Corrigidos', data=df, label='Bugs Corrigidos')
sns.lineplot(x='Dias', y='Tarefas Concludas', data=df, label='Tarefas Concludas')
plt.title('Progresso Dirio')
plt.xlabel('Dias')
plt.ylabel('Quantidade')
plt.legend()
plt.show()

# Total de Horas Trabalhadas e Mdia Diria de Horas Trabalhadas

total_horas = df['Horas Trabalhadas'].sum()
media_diaria_horas = df['Horas Trabalhadas'].mean()

print(f"Total de Horas Trabalhadas: {total_horas}")
print(f"Mdia Diria de Horas Trabalhadas: {media_diaria_horas:.2f}")

# Total de Bugs Corrigidos e Mdia Diria de Bugs Corrigidos

total_bugs = df['Bugs Corrigidos'].sum()
media_diaria_bugs = df['Bugs Corrigidos'].mean()

print(f"Total de Bugs Corrigidos: {total_bugs}")
print(f"Mdia Diria de Bugs Corrigidos: {media_diaria_bugs:.2f}")

# Total de Tarefas Concludas e Mdia Diria de Tarefas Concludas

total_tarefas = df['Tarefas Concludas'].sum()
media_diaria_tarefas = df['Tarefas Concludas'].mean()
print(f"Total de Tarefas Concludas: {total_tarefas}")
print(f"Mdia Diria de Tarefas Concludas: {media_diaria_tarefas:.2f}")

# Produtividade Diria (Tarefas Concludas por Hora)

df['Produtividade Diria'] = df['Tarefas Concludas'] / df['Horas Trabalhadas']

# Grfico de barras para Produtividade Diria
plt.figure(figsize=(8, 5))
sns.barplot(x='Dias', y='Produtividade Diria', data=df, color='skyblue')
plt.title('Produtividade Diria (Tarefas por Hora)')
plt.xlabel('Dias')
plt.ylabel('Produtividade')
plt.show()
# Concluses e Insights:
```
